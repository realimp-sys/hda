```python
# Experiment 5: Named Entity Recognition (NER) for Clinical Entities

# 1. Install required libraries
!pip install spacy med7

# 2. Import libraries
import spacy


# 3. Load the Med7 NLP model (clinical NER model)
nlp = spacy.load("en_core_med7_lg")

# 4. Sample clinical text
clinical_text = """
Patient presents with fever and cough for 3 days. Prescribed Amoxicillin 500mg thrice daily. 
Past history includes hypertension and type 2 diabetes mellitus. No known drug allergies.
Patient scheduled for MRI scan next week.
"""

doc = nlp(clinical_text)

# Extract entities
print("Named Entities in Clinical Text:\n")
for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}")

# Visualize entities
from spacy import displacy
displacy.render(doc, style="ent", jupyter=True)

```
