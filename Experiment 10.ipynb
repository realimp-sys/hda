```python
# ============================================================
# Experiment 10: Social Media Analytics using Survey Responses
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
survey_df = pd.read_csv('sentimentdataset.csv')

# Inspect columns
print("Columns in dataset:", survey_df.columns)

# 1. Distribution of posts by Platform
plt.figure(figsize=(6,4))
sns.countplot(x='Platform', data=survey_df)
plt.title("Posts by Platform")
plt.xticks(rotation=45)
plt.show()

# 2. Distribution of posts by Sentiment
plt.figure(figsize=(6,4))
sns.countplot(x='Sentiment', data=survey_df)
plt.title("Posts by Sentiment")
plt.show()

# 3. Average Likes and Retweets per Platform
platform_metrics = survey_df.groupby('Platform')[['Likes', 'Retweets']].mean().reset_index()
print("\nAverage Likes and Retweets per Platform:")
print(platform_metrics)

# Visualization
platform_metrics.plot(x='Platform', y=['Likes', 'Retweets'], kind='bar', figsize=(8,5))
plt.title("Average Likes and Retweets per Platform")
plt.ylabel("Count")
plt.show()

# 4. Top hashtags
# Flatten all hashtags into a list (assuming comma-separated hashtags)
hashtags_series = survey_df['Hashtags'].dropna().str.split(',').explode()
top_hashtags = hashtags_series.value_counts().head(10)
print("\nTop 10 Hashtags:")
print(top_hashtags)

# Visualization
top_hashtags.plot(kind='bar', figsize=(8,5))
plt.title("Top 10 Hashtags")
plt.ylabel("Frequency")
plt.show()

# 5. Temporal Analysis: Posts per Hour
posts_per_hour = survey_df.groupby('Hour').size()
posts_per_hour.plot(kind='bar', figsize=(8,5))
plt.title("Number of Posts per Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Posts")
plt.show()
```
