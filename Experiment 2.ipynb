```python
# Experiment 2: Data Analysis and Visualization on EHR

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization settings
sns.set(style='whitegrid', palette='muted', font_scale=1.1)

# 2. Load Dataset
DATA_PATH = 'diabetes.csv'  # Replace with your EHR dataset path
df = pd.read_csv(DATA_PATH)

# Quick overview
print(df.head())
print(df.info())
print(df.describe())

# 3. Descriptive Statistics
print("Mean values of numeric features:\n", df.mean())
print("Median values of numeric features:\n", df.median())
print("Standard deviation of numeric features:\n", df.std())

# 4. Univariate Analysis
# Histogram - Glucose
plt.figure(figsize=(8,5))
sns.histplot(df['Glucose'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose")
plt.ylabel("Count")
plt.show()

# Boxplot - BMI
plt.figure(figsize=(8,5))
sns.boxplot(x=df['BMI'], color='salmon')
plt.title("Boxplot of BMI")
plt.show()

# Density Plot - Age
plt.figure(figsize=(8,5))
sns.kdeplot(df['Age'], shade=True, color='green')
plt.title("Density Plot of Age")
plt.show()

# 5. Correlation Analysis
corr_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# 6. Bivariate Analysis
# Scatter plot - Age vs Glucose colored by Outcome
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Glucose', hue='Outcome', data=df)
plt.title("Age vs Glucose Level")
plt.xlabel("Age")
plt.ylabel("Glucose Level")
plt.show()

# Pairplot - Selected features
selected_features = ['Age', 'BMI', 'Glucose', 'BloodPressure', 'Outcome']
sns.pairplot(df[selected_features], hue='Outcome', palette='Set2')
plt.show()

# Violin plot - Glucose vs Outcome
plt.figure(figsize=(8,5))
sns.violinplot(x='Outcome', y='Glucose', data=df, palette='pastel')
plt.title("Glucose Distribution by Outcome")
plt.show()

# 7. Insights
# Example: Top 5 patients with highest Glucose
top_glucose = df.sort_values('Glucose', ascending=False).head(5)
print("Top 5 patients by Glucose:\n", top_glucose)
```
