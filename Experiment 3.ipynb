```python
# Experiment 3: Biomedical Image Preprocessing

# 1. Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure

# 2. Load Biomedical Image
# Replace 'biomedical_image.png' with your image file
image_path = 'image.png'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # load as grayscale
plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# 3. Resize Image
img_resized = cv2.resize(img, (256, 256))
plt.figure(figsize=(6,6))
plt.imshow(img_resized, cmap='gray')
plt.title("Resized Image")
plt.axis('off')
plt.show()

# 4. Normalization (scale pixel values to 0-1)
img_normalized = img_resized / 255.0
plt.figure(figsize=(6,6))
plt.imshow(img_normalized, cmap='gray')
plt.title("Normalized Image")
plt.axis('off')
plt.show()

# 5. Noise Reduction
# Gaussian Blur
img_gaussian = cv2.GaussianBlur(img_resized, (5,5), 0)
plt.figure(figsize=(6,6))
plt.imshow(img_gaussian, cmap='gray')
plt.title("Gaussian Blurred Image")
plt.axis('off')
plt.show()

# Median Filter
img_median = cv2.medianBlur(img_resized, 5)
plt.figure(figsize=(6,6))
plt.imshow(img_median, cmap='gray')
plt.title("Median Filtered Image")
plt.axis('off')
plt.show()

# 6. Histogram Equalization
img_hist_eq = cv2.equalizeHist(img_resized)
plt.figure(figsize=(6,6))
plt.imshow(img_hist_eq, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis('off')
plt.show()

# 7. Edge Detection
edges = cv2.Canny(img_resized, 100, 200)
plt.figure(figsize=(6,6))
plt.imshow(edges, cmap='gray')
plt.title("Edges Detected (Canny)")
plt.axis('off')
plt.show()

# 8. Optional: Contrast Stretching (using skimage)
img_rescale = exposure.rescale_intensity(img_resized, in_range='image', out_range=(0,255))
plt.figure(figsize=(6,6))
plt.imshow(img_rescale, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis('off')
plt.show()

# 9. Optional: Save Preprocessed Image
cv2.imwrite('biomedical_image_preprocessed.png', img_hist_eq)
print("Preprocessed image saved successfully!")
```
